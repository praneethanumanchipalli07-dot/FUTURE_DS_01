```python
#revenue trends
revenue_trends = df.groupby('Order Date')['Sales'].sum()
revenue_trends
```

```python
top_selling_product = df.groupby('Product Name')['Sales'].sum()
top_selling_product = top_selling_product.sort_values(ascending = False)
top_selling_product.head(10)
```

```python
high_categories = df.groupby('Category')['Sales'].sum()
high_categories
```

```python
region_performance = df.groupby('Region')['Sales'].sum()
region_performance
```

```python
import matplotlib.pyplot as plt
```

```python
high_categories.plot(kind = 'bar')
plt.title("Sales by Category")
plt.show()
```

```python
region_performance.plot(kind = 'bar')
plt.title("Regional Performance")
plt.show()
```

```python
top5 = top_selling_product.head(5)
```

```python
plt.pie(top5,autopct = '%1.1f%%')
plt.title("Top 5 Selling Products")
plt.legend(labels = top5.index)
plt.show()
```

```python
revenue_trends.plot(kind = 'line')
plt.title('Revenue Trends')
plt.show()
```
